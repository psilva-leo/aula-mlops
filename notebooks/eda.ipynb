```python
import os
import kagglehub
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Download Ames Housing Dataset

```python
path = kagglehub.dataset_download("shashanknecrothapa/ames-housing-dataset")

print("Path to dataset files:", path)
```

```python
os.listdir(path)
```

## Load dataset and show basic information

```python
df = pd.read_csv(os.path.join(path, "AmesHousing.csv"))
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
null_columns = df.columns[df.isnull().any()]

# Total number of rows
total_rows = len(df)

# Print information for each column with null values
for col in null_columns:
    num_nulls = df[col].isnull().sum()
    proportion_nulls = num_nulls / total_rows
    print(f"Column: {col}")
    print(f"  Type: {df[col].dtype}")
    print(f"  Total rows: {total_rows}")
    print(f"  Null values: {num_nulls}")
    print(f"  Proportion of nulls: {proportion_nulls:.2%}")
```

```python
df.isna().sum()
```

## Plot proportion of unique values if the max is greater than 30% (for visualization reason)

```python
for col in df.columns:
    value_counts = df[col].value_counts(dropna=False)
    proportions = value_counts / total_rows
    if max(proportions) < 0.3:
        continue

    plot_df = pd.DataFrame({'Value': value_counts.index, 'Proportion': proportions.values})

    plt.figure(figsize=(8, 6))
    sns.barplot(x='Value', y='Proportion', data=plot_df)
    plt.ylim(0, 1)
    plt.title(f'Proportion of Unique Values in Column "{col}"')
    plt.xlabel('Unique Values')
    plt.ylabel('Proportion')
    plt.xticks(rotation=45)
    plt.show()
```

## Correlation plots

```python
numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()

mask = (correlation_matrix.abs() > 0.7) & (correlation_matrix != 1)

high_corr_columns = [(correlation_matrix.columns[i], correlation_matrix.columns[j], correlation_matrix.iloc[i, j])
                     for i in range(len(correlation_matrix.columns))
                     for j in range(i+1, len(correlation_matrix.columns))
                     if mask.iloc[i, j]]

print("Total high correlation columns:", len(high_corr_columns))
for col1, col2, corr in high_corr_columns:
    print(f"Columns: {col1} and {col2} | Correlation: {corr:.2f}")

print("\nSale Price correlation matrix")
sale_price_corr = correlation_matrix['SalePrice']
high_corr_sale_price = sale_price_corr[abs(sale_price_corr) > 0.6].drop('SalePrice')
print(high_corr_sale_price)
print()
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()
```
